```python
warnings.simplefilter('ignore')
```

```python
#Membaca dataset dari direktori
df = pd.read_csv("5BreastCancer.csv")
df.head(35)
```

```python
feature_cols = ['Age', 'BMI', 'Glucose', 'Insulin', 'HOMA', 'Leptin', 'Adiponectin', 'Resistin', 'MCP.1']
X, y = df[feature_cols] , df.Classification
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2)
```

```python
ANN = MLPClassifier()
SVM = LinearSVC()
LR = LogisticRegression()
```

```python
ANN.fit (X_train, y_train)
ANN.score (X_train, y_train)
```

```python
SVM.fit (X_train, y_train)
SVM.score (X_train, y_train)
```

```python
LR.fit (X_train, y_train)
LR.score (X_train, y_train)
```

```python
ModelBagging = BaggingClassifier()
BC = BaggingClassifier(DecisionTreeClassifier(), max_samples= 0.5, 
                           max_features = 1.0, n_estimators = 20)
model = BC.fit(X_train, y_train)
y_pred = model.predict(X_test)
print ("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
